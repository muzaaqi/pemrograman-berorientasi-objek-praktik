```python
class Buku:
    def __init__(self, judul, penulis):
        self.judul = judul
        self.penulis = penulis
        self.dipinjam = False
    
    def pinjam(self):
        if not self.dipinjam:
            self.dipinjam = True
            print(f'Buku "{self.judul}" telah dipinjam.')
        else:
            print(f'Buku "{self.judul}" sedang dipinjam.')
    
    def kembalikan(self):
        if self.dipinjam:
            self.dipinjam = False
            print(f'Buku "{self.judul}" telah dikembalikan.')
        else:
            print(f'Buku "{self.judul}" belum dipinjam.')

buku1 = Buku("Belajar Python", "A. Programmer")
buku2 = Buku("Pemrograman Berorientasi Objek", "B. Developer")

buku1.pinjam()
buku1.kembalikan()
buku2.pinjam()
```

```python
class Menu:
    def __init__(self, nama, harga):
        self.nama = nama
        self.harga = harga
        self.tersedia = True
    
    def pesan(self):
        if self.tersedia:
            print(f'Menu "{self.nama}" telah dipesan.')
        else:
            print(f'Maaf, menu "{self.nama}" tidak tersedia.')

    def habis(self):
        if self.tersedia:
            self.tersedia = False
            print(f'Menu "{self.nama}" sekarang habis.')
        else:
            print(f'Menu "{self.nama}" sudah tidak tersedia.')
    
menu1 = Menu("Nasi Goreng", 15000)
menu2 = Menu("Mie Ayam", 12000)
menu3 = Menu("Es Teh", 5000)
menu4 = Menu("Jus Alpukat", 20000)
menu5 = Menu("Sate Ayam", 25000)

menu1.pesan()
menu2.habis()
menu2.pesan()
menu3.pesan()
menu4.habis()
menu5.pesan()
```

```python
class Anjing:
    def __init__(self, nama, umur):
        self.nama = nama
        self.umur = umur

    def menggonggong(self):
        print(f'{self.nama} sedang menggonggong.')

    def info(self):
        print(f'Anjing {self.nama} berumur {self.umur} tahun.')

anjing1 = Anjing("Buddy", 3)
anjing2 = Anjing("Max", 5)
anjing3 = Anjing("Molly", 2)
anjing4 = Anjing("Bella", 4)
anjing5 = Anjing("Charlie", 1)

anjing1.menggonggong()
anjing2.info()
anjing3.menggonggong()
anjing4.info()
anjing5.menggonggong()

```

```python
class Garis:
    def __init__(self, panjang):
        self.panjang = panjang

    def info(self):
        print(f'Garis dengan panjang {self.panjang} satuan.')
    
    def ubah_panjang(self, panjang_baru):
        self.panjang = panjang_baru
        print(f'Panjang garis diubah menjadi {self.panjang} satuan.')

garis1 = Garis(10)
garis2 = Garis(20)
garis3 = Garis(30)
garis4 = Garis(40)
garis5 = Garis(50)

garis1.info()
garis2.ubah_panjang(25)
garis3.info()
garis4.ubah_panjang(45)
garis5.info()
```

```python
class Pelajar:
    def __init__(self, nama, npm, jurusan):
        self.nama = nama
        self.npm = npm
        self.jurusan = jurusan

    def perkenalan(self):
        print(f'Halo, nama saya {self.nama}, NPM saya {self.npm} dan saya jurusan {self.jurusan}.')
    
    def ubah_jurusan(self, jurusan_baru):
        self.jurusan = jurusan_baru
        print(f'Jurusan saya telah diubah menjadi {self.jurusan}.')

pelajar1 = Pelajar("Andi", "123456789", "Informatika")
pelajar2 = Pelajar("Budi", "987654321", "Sistem Informasi")
pelajar3 = Pelajar("Citra", "456789123", "Teknik Komputer")
pelajar4 = Pelajar("Dewi", "789123456", "Teknologi Informasi")
pelajar5 = Pelajar("Eka", "321654987", "Ilmu Komputer")

pelajar1.perkenalan()
pelajar2.ubah_jurusan("Manajemen Informatika")
pelajar3.perkenalan()
pelajar4.ubah_jurusan("Rekayasa Perangkat Lunak")
pelajar5.perkenalan()
```
